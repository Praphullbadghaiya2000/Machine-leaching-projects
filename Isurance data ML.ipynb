```python
# data preprocessing 
# encoding the categorical features 
```

```python
# encoding sex column 
```

```python
ins_df.replace({'sex':{'male':0,'female':1}}, inplace=True)
```

```python
ins_df.sample(3)
```

```python
# encoding smoker column 
```

```python
ins_df.replace({'smoker':{'yes':0,'no':1}}, inplace=True)
```

```python
ins_df.sample(2)
```

```python
# encoding region column 
```

```python
ins_df.replace({'region':{'southeast':0,'southwest':1,'northeast':2,'northwest':3}}, inplace=True)
```

```python
ins_df
```

```python
# now we have all column as numerical input , here we will ML to predict the charges for insurance depends on various input column 
```

```python
# splitting the features and target 
```

```python
x=ins_df.iloc[:,0:6]
print(x)
```

```python
y=ins_df.iloc[:,-1]
print(y)
```

```python
# splitting the data into feature and tdrget 
```

```python
x_train , x_test , y_train,y_test = train_test_split(x,y,test_size=0.2 , random_state=2)
```

```python
print(x.shape ,x_train.shape,x_test.shape)
```

```python
# model training , linear regression , loading linear regression model 
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
# model evaluation
# prediction on training data
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
y_pred = reg.predict(x_train)
#Probability of each class
y_pred
```

```python
# r squared value 
```

```python
r2_train=metrics.r2_score(y_train, y_pred)
print("r2_score", r2_train)
```
